## Task 2 (Determine the optimal number of clusters)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist
```

```python
# importing the dataset
X = pd.read_csv('elbow.txt', sep = "\t", header= None, decimal = "," )
X.head()
```

```python
X.tail()
```

```python
# applying the elbow method
distortions = []
inertias = []
mapping1 = {}
mapping2 = {}
K = range(1, 10)

# building the model, calculating the values of distortion and inertia for each value of k in the given range
for k in K:
    kmeans = KMeans(n_clusters=k).fit(X)
    kmeans.fit(X)
    distortions.append(sum(np.min(cdist(X, kmeans.cluster_centers_,'euclidean'), axis=1)) / X.shape[0])
    inertias.append(kmeans.inertia_)
    mapping1[k] = sum(np.min(cdist(X, kmeans.cluster_centers_,'euclidean'), axis=1)) / X.shape[0]
    mapping2[k] = kmeans.inertia_
```

```python
# visualising the results
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('J(R)')
plt.title('The Elbow Method')
plt.grid()
plt.show()
```

The optimal number of clusters is 5 because starting from 5 the J(R) value does not decrease sharply as before.

```python
# repeating the search with silhouette score using the metrics module of the sklearn library
sil = {}
for k in range(2, 10):
    kmeans = KMeans(n_clusters = k).fit(X)
    labels = kmeans.labels_
    sil[k]= metrics.silhouette_score(X, labels)
```

```python
# visualising the results
plt.plot(sil.keys(), sil.values(), 'bx-')
plt.xlabel('k')
plt.ylabel('Silhouette score')
plt.grid()
plt.show()
```

By the silhouette method the optimal number is also 5. The two methods show the same result. 

## Task 3 (Hierarchical clustering)

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import calinski_harabasz_score, silhouette_samples, silhouette_score
```

```python
# importing the dataset and name columns
cnames = ['energy values', 'proteins', 'fats', 'calcium','iron']
df = pd.read_csv('nutrient.dat', header = None, names = cnames, sep = '   +')
```

```python
df.head()
```

```python
df.tail()
```

```python
# standardization of data because measures of data are different
df = StandardScaler().fit_transform(df)
```

```python
# calculating the silhouette-score for single agglomerative clustering to identify optimal number of clusters
J_single_sil = []
params = np.arange(2, 20)
for k in params:
    agg_single = AgglomerativeClustering(n_clusters=k, linkage='single', affinity='euclidean') 
    agg_single.fit(df)
    J_single_sil.append(silhouette_score(df, agg_single.labels_))
J_single_sil = np.array(J_single_sil)

# plotting the silhouette-score for single agglomerative clustering
plt.plot(params, J_single_sil, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette')
plt.grid()
plt.show()
```

```python
# calculating the silhouette-score for complete agglomerative clustering to identify optimal number of clusters
J_complete_sil = []
params = np.arange(2, 20)
for k in params:
    agg_complete = AgglomerativeClustering(n_clusters=k, linkage='complete', affinity='euclidean') 
    agg_complete.fit(df)
    J_complete_sil.append(silhouette_score(df, agg_complete.labels_))
J_complete_sil = np.array(J_complete_sil)

# plotting the silhouette-score for complete agglomerative clustering
plt.plot(params, J_complete_sil, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette')
plt.grid()
plt.show()
```

```python
# calculating the silhouette-score for average agglomerative clustering to identify optimal number of clusters
J_average_sil = []
params = np.arange(2, 20)
for k in params:
    agg_average = AgglomerativeClustering(n_clusters=k, linkage='average', affinity='euclidean') 
    agg_average.fit(df)
    J_average_sil.append(silhouette_score(df, agg_average.labels_))
J_average_sil = np.array(J_average_sil)

# plotting the silhouette-score for average agglomerative clustering
plt.plot(params, J_average_sil, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette')
plt.grid()
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage, cophenet
from scipy.spatial.distance import pdist

# linkage matrices 
Z_single = linkage(df, method='single', metric='euclidean')
Z_complete = linkage(df, method='complete', metric='euclidean')
Z_average = linkage(df, method='average', metric='euclidean')

c_single, coph_dists_single = cophenet(Z_single, pdist(df))
c_complete, coph_dists_complete = cophenet(Z_complete, pdist(df))
c_average, coph_dists_average = cophenet(Z_average, pdist(df))
```

```python
# selecting the threshold for dendrogram
def fancy_dendrogram(*args, **kwargs):
    max_d = kwargs.pop('max_d', None)
    if max_d and 'color_threshold' not in kwargs:
        kwargs['color_threshold'] = max_d
    annotate_above = kwargs.pop('annotate_above', 0)

    ddata = dendrogram(*args, **kwargs)

    if not kwargs.get('no_plot', False):
        plt.title('Hierarchical clustering dendrogram (cut at threshold)')
        plt.xlabel('Cluster size')
        plt.ylabel('Distance')
        for i, d, c in zip(ddata['icoord'], ddata['dcoord'], ddata['color_list']):
            x = 0.5 * sum(i[1:3])
            y = d[1]
            if y > annotate_above:
                plt.plot(x, y, 'o', c=c)
                plt.annotate("%.3g" % y, (x, y), xytext=(0, -5),
                             textcoords='offset points',
                             va='top', ha='center')
        if max_d:
            plt.axhline(y=max_d, c='k')
    return ddata
```

```python
# plotting the dendrogram using Z_single linkage matrix
plt.figure(figsize=(25, 10))
plt.title('Hierarchical clustering dendrogram (Single)')
plt.ylabel('Distance')
dendrogram(
    Z_single, # linkage matrix to use
    truncate_mode='lastp',  # show last clusters
    p=10,  # show 15 clusters
    show_leaf_counts=True,  # show the objects in leaf nodes
    leaf_rotation=90., # the angle to rotate the leaf labels
    leaf_font_size=12., # the font size on leaf labels
    show_contracted=True,  # show the heights of non-singleton nodes contracted into a leaf node
)
plt.show()
```

```python
# plotting the cut at a threshold hierarchical clustering dendrogram (single)
max_d = 3
plt.figure(figsize=(25, 10))
fancy_dendrogram(
    Z_single,
    truncate_mode='lastp',
    p=12,
    leaf_rotation=90.,
    leaf_font_size=12.,
    show_contracted=True,
    annotate_above=8,
    max_d=max_d,
)
plt.show()
```

```python
# plotting the dendrogram using Z_average linkage matrix
plt.figure(figsize=(25, 10))
plt.title('Hierarchical clustering dendrogram (Average)')
plt.ylabel('Distance')
dendrogram(
    Z_average, # linkage matrix to use
    truncate_mode='lastp', # show last clusters
    p=15, # show 15 clusters
    show_leaf_counts=True,  # show the objects in leaf nodes
    leaf_rotation=90., # the angle to rotate the leaf labels
    leaf_font_size=12., # the font size on leaf labels
    show_contracted=True,  # show the heights of non-singleton nodes contracted into a leaf node
)
plt.show()
```

```python
# plotting the cut at a threshold hierarchical clustering dendrogram (average)
max_d = 3
plt.figure(figsize=(25, 10))
fancy_dendrogram(
    Z_average,
    truncate_mode='lastp',
    p=12,
    leaf_rotation=90.,
    leaf_font_size=12.,
    show_contracted=True,
    annotate_above=8,
    max_d=max_d,
)
plt.show()
```
